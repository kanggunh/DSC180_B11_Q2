```python
from docling.document_converter import DocumentConverter
```

```python
##Convert  1.pdf

path_1 = "data/pdfs/1.pdf"
converter = DocumentConverter()
result = converter.convert(path_1)
result_markdown = result.document.export_to_markdown()
print(result_markdown)

with open('txt_pdf1.txt', 'w', encoding='utf-8') as f:
    f.write("\n\nMarkdown Conversion: \n")
    f.write(result_markdown)
```
