```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('data/data_expanded_SMILES.csv')
```

```python
df.head()
```

```python
unparsed = df[df['passivating_molecule_SMILES'].isna()]
```

```python
unparsed
```

```python
def get_smiles(molecule_name):
    base_url = "https://opsin.ch.cam.ac.uk/opsin/"
    smiles_url = base_url + molecule_name + ".smi"
    r = requests.get(smiles_url)
    return r.text if r.status_code == 200 else None
```

```python
get_smiles("phenylethylammonium lead iodide")
```

```python
unparsed["api_parsed"] = unparsed["passivating_molecule"].apply(get_smiles)
```
